# feature selection

```python
# Heatmap
dataset.corr()
sns.heatmap(dataset.corr(), annot = True)
plt.show()
```

```python

X = dataset.iloc[:, [1,2,3,4]].values
Y = dataset.iloc[:, [5]].values
```

```python
# Splitting X and Y
#Controls the shuffling applied to the data before applying the split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 42 )
```

```python
# Checking dimensions
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("Y_train shape:", Y_train.shape)
print("Y_test shape:", Y_test.shape)
```

```python
print(X_train)
```

#  MACHINE LEARNING ALGORITHMS

# Support Vector Classifier Algorithm

```python
# Support Vector Classifier Algorithm
from sklearn.svm import SVC
svc = SVC(kernel = 'linear', random_state = 42)
svc.fit(X_train, Y_train)
```

# Decision tree Algorithm

```python
# Decision tree Algorithm
from sklearn.tree import DecisionTreeClassifier
dectree = DecisionTreeClassifier(criterion = 'entropy', random_state = 42)
dectree.fit(X_train, Y_train)
```

# Random forest Algorithm

```python
# Random forest Algorithm
from sklearn.ensemble import RandomForestClassifier
ranfor = RandomForestClassifier(n_estimators = 11, criterion = 'entropy', random_state = 42)
ranfor.fit(X_train, Y_train)
```

# Making predictions on test dataset

```python

Y_pred_svc = svc.predict(X_test)
Y_pred_dectree = dectree.predict(X_test)
Y_pred_ranfor = ranfor.predict(X_test)
```

 # Model Evaluation

```python
# Evaluating using accuracy_score metric
from sklearn.metrics import accuracy_score
```

```python

accuracy_svc = accuracy_score(Y_test, Y_pred_svc)
accuracy_dectree = accuracy_score(Y_test, Y_pred_dectree)
accuracy_ranfor = accuracy_score(Y_test, Y_pred_ranfor)
```

```python
# Accuracy on test set

print("Support Vector Classifier: " + str(accuracy_svc * 100))
print("Decision tree: " + str(accuracy_dectree * 100))
print("Random Forest: " + str(accuracy_ranfor * 100))
```

```python

c=accuracy_svc * 100
d=accuracy_dectree * 100
e=accuracy_ranfor * 100
```

```python
scores = [c,d,e]
algorithms = ["Support Vector Machine","Decision Tree","Random Forest"]    

for i in range(len(algorithms)):
    print("The accuracy score achieved using "+algorithms[i]+" is: "+str(scores[i])+" %")
```

```python
sns.set(rc={'figure.figsize':(8,8)})
plt.xlabel("Algorithms")
plt.ylabel("Accuracy score")

sns.barplot(algorithms,scores)
```

```python

import joblib 
  
# Save the model as a pickle in a file 
joblib.dump(ranfor, r'Models\ranfor_model1.pkl') 
  
# Load the model from the file 
ranfor_from_joblib = joblib.load(r'Models\ranfor_model1.pkl')  
  
# Use the loaded model to make predictions 
ranfor_from_joblib.predict(X_test)
```

# Confusion Matrix

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred_ranfor)
```

```python
# Heatmap of Confusion matrix
plt.figure(figsize = (10,10))

sns.heatmap(pd.DataFrame(cm), annot=True)
```
